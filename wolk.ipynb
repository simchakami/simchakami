```python
import random
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
def walk_random_1D(steps):
    walk =[0]
    position = 0
    for i in range (1,steps):
        position += random.choice([1,-1])
        walk.append(position)
    return walk

    
def wolk_plt_1D(walk):
    plt.plot(walk)
    plt.xlabel("Time")
    plt.ylabel("position")
    plt.grid()
    plt.show()

wolk_plt_1D(walk_random_1D(100))


    
```

```python
def walk_random_2d(steps):
    walk_x = [0]
    walk_y = [0]

    possition_x = 0
    possition_y = 0
    diractions = ["East", "West", "North", "South"]
    for i in range(1,steps):
        possition = random.choice(diractions)
        if possition == "East":
            possition_x += 1
        elif possition == "West":
            possition_x -= 1
        elif possition == "North":
            possition_y += 1
        else:
            possition_y -= 1
        walk_x.append(possition_x)
        walk_y.append(possition_y)
    print(walk_x,walk_y)
    return walk_x,walk_y    
        
def walk_plt_2d(walk):
    plt.plot(walk[0],walk[1],marker = "o")
    plt.xlabel("Position in East - West diraction")
    plt.ylabel("Position in Nort - South diraction")
    plt.scatter(walk[0][0],walk[1][0],color = "green",linewidths=10)
    plt.scatter(walk[0][-1],walk[1][-1],color = "red",linewidths=10)
    plt.grid()
    plt.show()

walk_plt_2d(walk_random_2d(100))
```

```python
def walk_random_3d(steps):
    walk_x = [0]
    walk_y = [0]

    possition_x = 0
    possition_y = 0
    diractions = ["East", "West", "North", "South"]
    for i in range(1,steps):
        possition = random.choice(diractions)
        if possition == "East":
            possition_x += 1
        elif possition == "West":
            possition_x -= 1
        elif possition == "North":
            possition_y += 1
        else:
            possition_y -= 1
        walk_x.append(possition_x)
        walk_y.append(possition_y)
   
    print(walk_x,walk_y)
    return walk_x,walk_y    
 
def walk_plt_3d(walk):
    Time = list(range(len (walk[0])))
    ax = plt.figure().add_subplot(111,projection='3d')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    plt.title("random walw 3D",color="red")
    ax.plot(walk[0],walk[1],Time,marker = "o")
    ax.scatter(walk[0][0],walk[1][0],Time[0],color = "green",linewidths=10)
    ax.scatter(walk[0][-1],walk[1][-1],Time[-1],color = "red",linewidths=10)
    plt.grid()
    plt.show()

walk_plt_3d(walk_random_3d(10))
```
